```python
class Puzzle:
    def __init__(self,my_list,goal_list):
      self.current_list=my_list
      self.end_list=goal_list
      self.empty_tile=self.getEmpty()
      self.parent=None
      self.current_state = self.current_list
    def down(self):
      if self.empty_tile-3>=0:
          self.current_state[self.empty_tile]=self.current_state[self.empty_tile-3]
          self.current_state[self.empty_tile - 3]=0
          self.empty_tile=self.empty_tile-3
          return True
      else:
          print("Invalid move")
          return False
    def up(self):
      if self.empty_tile-5<=0:
          self.current_state[self.empty_tile]=self.current_state[self.empty_tile+3]
          self.current_state[self.empty_tile+3]=0
          self.empty_tile=self.empty_tile+3
          return True
      else:
          print("Invalid move")
          return False
    def right(self):
      if self.empty_tile!=2 and self.empty_tile!=5 and self.empty_tile!=8:
          self.current_state[self.empty_tile], self.current_state[self.empty_tile+1] = self.current_state[self.empty_tile+1], self.current_state[self.empty_tile]
          self.empty_tile=self.empty_tile+1
          return True
      else:
          print("Invalid move")
          return False
    def left(self):
      if self.empty_tile!=0 and self.empty_tile!=3 and self.empty_tile!=6:
          self.current_state[self.empty_tile], self.current_state[self.empty_tile-1] = self.current_state[self.empty_tile-1], self.current_state[self.empty_tile]
          self.empty_tile=self.empty_tile-1
          return True
      else:
        print("Invalid move")
        return False
    def getEmpty(self):
      for i in range(len(self.current_list)):
          if self.current_list[i]==0:
              return i
    def display(self):
      for row in self.current_list:
          print(row)
    def IsGoalReached(self):
      for i in range(len(self.current_list)):
        if self.current_list[i]!=self.end_list[i]:
          return False
      print("Goal Reached")
      return True
    def compare(self,p2):
      for i in range(len(self.current_list)):
        if self.current_list[i]!=p2.current_list[i]:
          return False
      return True
    def Checkopenclosedlist(self, p, open_stack, closed_list):
      for node in closed_list:
        if node.compare(p):
           return True
      for node in open_stack:
        if node.compare(p):
          return True
      return False
    def move(self,direction):
      if direction=="up":
        return self.up()
      elif direction=="down":
        return self.down()
      elif direction=="left":
        return self.left()
      elif direction=="right":
        return self.right()
def dfs(start, goal):
    open_stack = [start]
    closed_set = set()
    while open_stack:
        node = open_stack.pop()
        node.display()
        if node.IsGoalReached():
            return node
        closed_set.add(node)
        for direction in ["up", "down", "left", "right"]:
            new_node = Puzzle(node.current_list.copy(), goal)
            if new_node.move(direction) and new_node not in closed_set:
                if not new_node.Checkopenclosedlist(new_node, open_stack, closed_set):
                    new_node.parent = node
                    open_stack.append(new_node)
    return None
initial_state = [1, 2, 3, 8, 0, 4, 7, 6, 5]
goal_state = [2, 8, 1, 0, 4, 3, 7, 6, 5]
start_puzzle = Puzzle(initial_state, goal_state)
bfs(start_puzzle, goal_state)
```

```python
class WaterJug:
    def __init__(self, jug1, jug2, target):
        self.jug1_cap = jug1
        self.jug2_cap = jug2
        self.target = target
        self.jug1 = 0
        self.jug2 = 0
        self.parent = None

    def fill1(self):
        self.jug1 = self.jug1_cap

    def fill2(self):
        self.jug2 = self.jug2_cap

    def empty1(self):
        self.jug1 = 0

    def empty2(self):
        self.jug2 = 0

    def pour1to2(self):
        transfer = min(self.jug1, self.jug2_cap - self.jug2)
        self.jug1 -= transfer
        self.jug2 += transfer

    def pour2to1(self):
        transfer = min(self.jug2, self.jug1_cap - self.jug1)
        self.jug2 -= transfer
        self.jug1 += transfer

    def display(self):
        print(f"Jug1: {self.jug1}L, Jug2: {self.jug2}L")

    def IsGoalReached(self):
        return self.jug1 == self.target

    def Checkopenclosedlist(self, open_list, closed_list):
        return any(node.jug1 == self.jug1 and node.jug2 == self.jug2 for node in open_list + closed_list)

def dfs(jug1_cap, jug2_cap, target):
    start = WaterJug(jug1_cap, jug2_cap, target)
    open_stack = [start]
    closed_set = set()

    while open_stack:
        node = open_stack.pop()
        node.display()

        if node.IsGoalReached():
            return node

        closed_set.add((node.jug1, node.jug2))

        for action in ["fill1", "fill2", "empty1", "empty2", "pour1to2", "pour2to1"]:
            new_node = WaterJug(jug1_cap, jug2_cap, target)
            new_node.jug1, new_node.jug2 = node.jug1, node.jug2
            getattr(new_node, action)()

            if (new_node.jug1, new_node.jug2) not in closed_set:
                new_node.parent = node
                open_stack.append(new_node)

    return None


jug1_capacity = 4
jug2_capacity = 3
target_amount = 2
print("\nDFS Solution:")
solution_dfs = dfs(jug1_capacity, jug2_capacity, target_amount)
```

```python
#Travelling Salseman Problem
class TSP:
    def __init__(self, distances, starting_city):
        self.distances = distances
        self.starting_city = starting_city - 1
        self.current_path = [self.starting_city]
        self.minimum_distance = float('inf')
        self.best_path = []
        self.parent = None
    def calculate_distance(self, path):
        distance = 0
        for i in range(len(path) - 1):
            distance += self.distances[path[i]][path[i + 1]]
        distance += self.distances[path[-1]][path[0]]
        return distance
    def is_goal_reached(self):
        return len(self.current_path) == len(self.distances)
    def move(self, next_city):
        self.current_path.append(next_city)
        return True
    def check_open_closed_list(self, open_stack, closed_list):
        for node in closed_list:
            if node.current_path == self.current_path:
                return True
        for node in open_stack:
            if node.current_path == self.current_path:
                return True
        return False
    def bfs(self):
        open_stack = [self]
        closed_list = []
        while open_stack:
            node = open_stack.pop(0)
            if node.is_goal_reached():
                total_distance = node.calculate_distance(node.current_path)
                if total_distance < self.minimum_distance:
                    self.minimum_distance = total_distance
                    self.best_path = node.current_path
                continue
            closed_list.append(node)
            for city in range(len(self.distances)):
                if city not in node.current_path:
                    new_node = TSP(node.distances, node.starting_city + 1)
                    new_node.current_path = node.current_path.copy()
                    new_node.parent = node
                    if new_node.move(city):
                        if not new_node.check_open_closed_list(open_stack, closed_list):
                            open_stack.append(new_node)
distance = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]
starting_city = int(input("Starting city: "))
initial_state = TSP(distance, starting_city)
initial_state.bfs()
print("Optimal path:", [city + 1 for city in initial_state.best_path])
print("Minimum distance:", initial_state.minimum_distance)
```
